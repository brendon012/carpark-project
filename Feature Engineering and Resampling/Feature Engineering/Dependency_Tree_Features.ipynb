```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import spacy
import numpy as np
nlp = spacy.load("en_core_web_sm")


train_data = pd.read_csv("in_domain_train.tsv", sep="\t", header=None)
train_sentences = train_data.iloc[:, 3]
train_labels = train_data.iloc[:, 1]

test_data = pd.read_csv("in_domain_dev.tsv", sep="\t", header=None)
test_sentences = test_data.iloc[:, 3]
test_labels = test_data.iloc[:, 1]

train_num_dependencies = []
train_avg_tree_depth = []

for sentence in train_sentences:
    doc = nlp(sentence)

    # Calculate the number of dependencies
    num_deps = sum(1 for token in doc if token.head != token)
    train_num_dependencies.append(num_deps)

    # Calculate the average tree depth
    tree_depths = [token.dep_ for token in doc]
    avg_depth = sum(len(dep.split("-")) for dep in tree_depths) / len(tree_depths)
    train_avg_tree_depth.append(avg_depth)


test_num_dependencies = []
test_avg_tree_depth = []

for sentence in test_sentences:
    doc = nlp(sentence)

    num_deps = sum(1 for token in doc if token.head != token)
    test_num_dependencies.append(num_deps)

    tree_depths = [token.dep_ for token in doc]
    avg_depth = sum(len(dep.split("-")) for dep in tree_depths) / len(tree_depths)
    test_avg_tree_depth.append(avg_depth)


vectorizer = CountVectorizer(ngram_range=(1, 3))


train_vec = vectorizer.fit_transform(train_sentences)
test_vec = vectorizer.transform(test_sentences)

train_features = train_vec.toarray()
train_features = np.hstack((train_features, np.array(train_num_dependencies).reshape(-1, 1), np.array(train_avg_tree_depth).reshape(-1, 1)))
test_features = test_vec.toarray()
test_features = np.hstack((test_features, np.array(test_num_dependencies).reshape(-1, 1), np.array(test_avg_tree_depth).reshape(-1, 1)))


lr = LogisticRegression(max_iter=1000, random_state=42)

lr.fit(train_features, train_labels)


test_pred = lr.predict(test_features)

test_accuracy = metrics.accuracy_score(test_labels, test_pred)
test_precision = metrics.precision_score(test_labels, test_pred)
test_recall = metrics.recall_score(test_labels, test_pred)

print(f"Test Accuracy: {test_accuracy}")
print(f"Test Precision: {test_precision}")
print(f"Test Recall: {test_recall}")
```
